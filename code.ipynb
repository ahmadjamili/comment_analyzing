##### using regex we are gonna stem the words in farsi , removing its stopwords and those alphabets that have been repeated more than once in a word

```python
import re
with open('C:/Users/ahmad/OneDrive/Desktop/persian.txt','r',encoding='utf-8') as f:
    stopwords_farsi=f.read().splitlines()
#stopwords_farsi.append("!")
stopwords_farsi.remove('کم')
stopwords_farsi.remove('اصلا')
stopwords_farsi.remove('خوب')
stopwords_farsi.remove('نبود')
f.close()
def stemming(content):
    content=re.sub('[<>#.:()"\'@$%؟^&*-+]',' ',content)
    content=content.split()
    pattern=r'(.)\1+'
    content=[re.sub(pattern,r'\1',word) and re.sub('!',' بد ' , word) for word in content if not word in stopwords_farsi]
    content=' '.join(content)
    content=content.replace('u200c','')
    return content
    
```

```python
dataset['comment1']=dataset['comment'].apply(stemming)
```

##### using the downloaded file consists of good attributes and bad attributes to label the texts using rule based way 

```python
import pandas as pd
Persentdataset=pd.read_excel(open("C:/Users/ahmad/OneDrive/Desktop/farsi nlp/PerSent1.xlsx",'rb'))
```

```python
Persentdataset
```

```python
farsi_sentiment={}
for i in range(len(Persentdataset['Words'].values)):
    farsi_sentiment[Persentdataset['Words'].values[i]]=float(Persentdataset['Polarity'].values[i])
```

```python
farsi_sentiment
```

```python
farsi_sentiment['عالی']=0.5
farsi_sentiment['نبود']=-0.7
```

```python
list(farsi_sentiment.keys())
```

```python
def PerSentanalyzing(text):
    content=text.split()
    bool=0
    values=0
    for i in content:
        if i in list(farsi_sentiment.keys()):
            values+=farsi_sentiment[i]
            bool=1
    if bool==0:
        values=np.nan
        
    return values
```

```python
dataset['PerSentpolarity']=dataset['comment1'].apply(PerSentanalyzing)
```

```python
dataset
```

```python
dataset['PerSentpolarity'].isna().sum()
```

```python
dataset=dataset.fillna('-0.6')
```

```python
dataset
```

```python
def final_label(num):
    num=float(num)
    if num>0:
        return 1
    if num<=0:
        return 0
```

```python
dataset['label']=dataset['PerSentpolarity'].apply(final_label)
```

#####  deviding the dataset into two trainset and testset

```python
x=train_dataset['comment1'].values
```

```python
y=train_dataset['label'].values
```

##### vectorizing out texts

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(min_df=4)
x = vectorizer.fit_transform(x).toarray()
```

```python
x.shape
```

##### as you saw in two previous step we figured out that we have a unbalanced data

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(sampling_strategy='majority')
x_train1, y_train1 = undersample.fit_resample(x, y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_train1,y_train1,test_size=0.50)
```

```python
x_train.shape
```

##### first we use the classic model in machine learning and for this binary classification i decided to use the logistic regression 

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
from sklearn.metrics import f1_score,confusion_matrix
prediction=model.predict(x_test)
print(f'accuracy score: {round(f1_score(prediction,y_test),3)*100}')
print(confusion_matrix(prediction,y_test))
```

```python
from sklearn.metrics import roc_auc_score
auc_score = roc_auc_score(prediction, y_test, multi_class='ovr')
auc_score
```
